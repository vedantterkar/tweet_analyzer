## Data Visualization
* Read Prepared Data
* Get Range of Data; ie, spread of data
* Get Details of spread of data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Read Prepared Data

```python
data = pd.read_json("data/preparedtweets.json")
data.head()
```

### Get Range of Data

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.hist(data["target"], edgecolor='black');
ax.set(title="Data Range", xlabel='value', ylabel='count');
fig.savefig("data/images/hist.png");
```

### Get Spread of Data

```python
fig1, ax1 = plt.subplots(figsize=(30,10))
ax1.scatter(data.index, data["target"])
ax1.set(title="Data Distribution", xlabel="index", ylabel="value");
fig1.savefig("data/images/dist.png");
```
